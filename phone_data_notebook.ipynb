```python
import pandas as pd
import numpy as np
import xlrd
import json
from datetime import datetime
```

```python
raw_df = pd.read_excel(
    "data/P641Lifecyclereport2020072318165587.xls",
    skiprows=7
)

raw_df.dropna(axis=1, how='all', inplace=True)
```

```python
raw_df.head()
```

```python
df_list = np.split(raw_df, raw_df[raw_df.iloc[:, 0] == "Time:"].index)
```

```python
for df in df_list:
    
    if not df.empty:
        call_info = df.iloc[0]
        call_info.dropna(inplace=True)
        df = df.iloc[1:]
        df.dropna(axis=1, how='all', inplace=True)
        header = df.iloc[0]
        df = df.iloc[1:]
        df.columns = header
        
        df.reset_index(drop=True, inplace=True)
        df.columns.name = None
        
        df['Call_Direction'] = call_info[5]
        df['Phone_Number'] = call_info[7]
        df['Location'] = call_info[9]
        
        print(df.head())
```
